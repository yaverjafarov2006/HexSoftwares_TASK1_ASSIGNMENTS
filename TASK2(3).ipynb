```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

# ----------------------------
# 1. SAMPLE DATA (əgər dataset yoxdursa)
# ----------------------------
data = {
    "Movie": ["A", "B", "C", "D", "E", "F", "G", "H"],
    "Genre": ["Action", "Drama", "Comedy", "Action", "Drama", "Comedy", "Action", "Drama"],
    "Rating": [8.5, 7.2, 6.8, 9.0, 7.5, 6.5, 8.8, 7.9]
}

df = pd.DataFrame(data)

# ----------------------------
# 2. CLEANING
# ----------------------------
df = df.dropna()

# ----------------------------
```

```python
# 3. SUMMARY STATISTICS
# ----------------------------
mean_rating = df["Rating"].mean()
median_rating = df["Rating"].median()
mode_rating = df["Rating"].mode()[0]

print("=== SUMMARY ===")
print(f"Mean   : {mean_rating:.2f}")
print(f"Median : {median_rating:.2f}")
print(f"Mode   : {mode_rating:.2f}")

# ----------------------------
# 4. VISUALIZATION
# ----------------------------

# Histogram
plt.figure()
plt.hist(df["Rating"])
plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()

# Boxplot
plt.figure()
sns.boxplot(x=df["Rating"])
plt.title("Boxplot of Ratings")
plt.show()

```

```python
# ----------------------------
# 5. TOP RATED MOVIES
# ----------------------------
top_movies = df.sort_values(by="Rating", ascending=False)
print("\n=== TOP MOVIES ===")
print(top_movies)

# Top genres
top_genres = df.groupby("Genre")["Rating"].mean().sort_values(ascending=False)
print("\n=== TOP GENRES ===")
print(top_genres)

# ----------------------------
# 6. SIMPLE DASHBOARD (plots)
# ----------------------------

plt.figure(figsize=(10,5))

# Bar plot (genre avg rating)
plt.subplot(1,2,1)
top_genres.plot(kind="bar")
plt.title("Average Rating by Genre")

# Scatter plot
plt.subplot(1,2,2)
plt.scatter(df.index, df["Rating"])
plt.title("Ratings Scatter")

plt.tight_layout()
plt.show()
```
